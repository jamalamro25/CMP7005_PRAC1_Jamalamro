```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

combine the csv files into one dataset sheet using one dataframe method

```python
drive_path = '/content/drive/MyDrive/Programming for Data Analysis/Assigment/Assessment Data-20251028'
```

```python
dataframes = []
for filename in os.listdir(drive_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(drive_path, filename)
        df = pd.read_csv(os.path.join(drive_path, filename))
        dataframes.append(df)
```

```python
df1 = pd.concat(dataframes, ignore_index=True)
```
