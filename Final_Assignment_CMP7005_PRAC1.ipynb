```python
def missing_value_table(df):
        mis_val = df.isnull().sum()
        mis_val_per = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_per], axis=1)
        mis_val_table = mis_val_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table = mis_val_table[mis_val_table.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False)
        return mis_val_table
missing_values = missing_value_table(df)
display(missing_values.style.background_gradient(cmap='Reds'))

```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

df.replace(["", " ", "-" , "--","NA", "N/A", "na", "null", "None"], np.nan, inplace=True)

for col in df.columns:
    if df[col].dtype == 'object' and col not in ['City', 'AQI_Bucket', 'Date']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print('-' * 50)

special_numeric_cols = {"Xylene", "AQI", "PM10", "Toluene", "Benzene", "NH3"}
for col in ["PM10", "Toluene", "Benzene"]:
    if col in df.columns:
        med = df[col].median(skipna=True)
        df[col]= df[col].fillna(med) # Corrected inplace=True usage
        print(f"{col} column filled with median value")
        print('-' * 50)

if "NH3" in df.columns and "City" in df.columns and df["NH3"].isna().any():
    before = df["NH3"].isna().sum()

    predicators = [c for c in df.select_dtypes(include=[np.number]).columns if c not in ["NH3", "Xylene", "AQI"]]
    for city, grp in df.groupby("City"):
        train = grp.dropna(subset=predicators + ["NH3"]) # Ensure NH3 is not NaN for training
        test = grp[grp["NH3"].isna() & grp[predicators].notna().all(axis=1)] # Fixed: Added closing bracket ']'
        if len(train) >= 10 and len(test) > 0:
            model = LinearRegression()
            model.fit(train[predicators], train["NH3"])
            df.loc[test.index, "NH3"] = model.predict(test[predicators])

    df["NH3"] = df.groupby("City")["NH3"].transform(lambda x: x.fillna(x.median()))
    df["NH3"] = df["NH3"].fillna(df["NH3"].median()) # Corrected inplace=True usage
    after = df["NH3"].isna().sum()
    print(f"NH3 regression filled {before - after} values (with fallbacks).")

for col in df.select_dtypes(include=[np.number]).columns:
    if col not in special_numeric_cols and df[col].isna().any():
        mean_val = df[col].mean(skipna=True)
        df[col] = df[col].fillna(mean_val) # Corrected inplace=True usage
        print(f"{col}: column filled with mean value")

for col in ["Xylene", "AQI"]:
    if col in df.columns and df[col].isna().any():
        mean_val = df[col].mean(skipna=True)
        df[col] = df[col].fillna(mean_val) # Corrected inplace=True usage
        print(f"{col}: column filled with mean value")
print('-' * 50)

categorical_cols = ["City", "AQI_Bucket", "Date", "Year", "Month"]

for col in categorical_cols:
    if col in df.columns and df[col].isna().any():
        mode_val = df[col].mode()
        if not mode_val.empty:
            df[col] = df[col].fillna(mode_val[0]) # Corrected inplace=True usage
            print(f"{col}: column filled with mode value")
        else:
            print(f"{col}: no mode value found")

print(df.isnull().sum().sort_values(ascending=False).head(20)) # Fixed: Added closing parenthesis ')'
```
